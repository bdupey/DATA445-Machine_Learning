```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt; plt.style.use('ggplot')
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
heart=pd.read_csv('framingham.csv')
heart.head()
```

```python
heart.describe()
```

```python
heart=heart.dropna()
```

```python
sns.scatterplot(data=heart,x='heartRate',y='BMI',color='steelblue');
```

```python
# From the above chart, we don't see any relationship between BMI and heartRate.
```

```python
sns.scatterplot(data=heart,x='glucose',y='BMI',color='steelblue');
```

```python
# From the above chart, we don't see any relationship between BMI and glucose.
```

```python
X=heart[['age','currentSmoker','totChol','BMI','heartRate']]
Y=heart['TenYearCHD']

# Building the model
logit_md=LogisticRegression().fit(X,Y)

# Predicting the likelihood of TenYearCHD
new_obs=[[50,1,250,27,85]]

logit_md.predict_proba(new_obs)
```

```python
# From the above results, we can see that there is a 82% likelihood of not having heart disease. 
```
